```python
import pandas as pd
import reverse_geocoder as rg 
import pprint 
```

```python
coor_df = pd.read_csv('facility_latlon_data.csv')
coor_df
```

```python
def reverseGeocode(coordinates): 
    result = rg.search(coordinates)
    return (result)

if __name__=="__main__": 

    # Coordinates tuple.Can contain more than one pair.
    coordinates =list(zip(coor_df['FacilityLatitude'],coor_df['FacilityLongitude'])) # generates pair of (lat,long)
    data = reverseGeocode(coordinates)

    coor_df['name'] = [city['name'] for city in data]
    coor_df['admin1'] = [city['admin1'] for city in data]

coor_df.to_csv("data_appended.csv") # write to csv # result will be saved to data_appended.csv
```
